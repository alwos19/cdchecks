## Prueba de hipótesis
* Revisaremos la estadistica del caso de violación

```python
sample_size = 100
eligible_population = [0.26, 0.74]
```

```python
sample_proportions(sample_size, eligible_population).item(0)
```

```python
sample_proportions(sample_size, eligible_population).item(0)
```

```python
def one_simulated_count():
    return sample_size * sample_proportions(sample_size, eligible_population).item(0)
```

```python
one_simulated_count
```

```python
counts = make_array()
repetitions = 10000
for i in np.arange(repetitions):
    counts = np.append(counts, one_simulated_count())
```

```python
counts
```

```python
Table().with_column('Count in a Random Sample', counts).hist(bins = np.arange(5.5, 46.6, 1))
```
